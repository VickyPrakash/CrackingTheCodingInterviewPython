```python
class Node:
    def __init__(self, dataval = None):
        self.data = dataval;
        self.next = None;

class Stack:
    def __init__(self):
        self.head = None;
    
    def push(self, value):
        newNode = Node(value);
        newNode.next = self.head;
        self.head = newNode;
    
    def pop(self):
        if(self.head == None):
            print("Empty List");
        else:
            top = self.head.next;
            print("Item popped: ", self.head.data);
            self.head = top;
            
    def peek(self):
        if(self.head == None):
            print("Empty List");
        else:
            print(self.head.data);
    def isEmpty(self):
        if(self.head == None):
            print("Empty List");
            return True;
        else: 
            print("Not Empty List");
            return False;
            
    def printStack(self):
        n = self.head;
        print(n.data);
        n = n.next;
        while(n.next != None):
            print(n.data);
            n = n.next;
        print(n.data);
        
def SortStack(stack1):
    stack2 = Stack();
    n = stack1.head;
    stack2.push(n.data);
    while(n.next != None):
        n = n.next;
        a = n.data;
        m = stack2.head;
        if(a >= m.data):
            stack2.push(a);
            print("a: ", a);
        else:
            k = 0;
            while(a < stack2.head.data):
                stack1.push(stack2.head.data);
                stack2.pop();
                k += 1;
                if(stack2.isEmpty()):
                    #stack2.push(a);
                    break;
            stack2.push(a);
            for i in range(k):
                stack2.push(stack1.head.data);
                stack1.pop();
    print("Sorted Stack");
    stack2.printStack();
    stack1.head = None;
    while(stack2.head != None):
        stack1.push(stack2.head.data);
        stack2.pop();
    
            
stack1 = Stack();
stack1.isEmpty();
stack1.peek();
stack1.pop();
stack1.push(13);
stack1.push(4);
stack1.push(12);
stack1.push(8);
stack1.printStack();
print("Sorted Stack")
SortStack(stack1);
print("Done")
stack1.peek();
#stack1.pop();
print("Stack in increasing order");
stack1.printStack();
stack1.isEmpty();
```

```python
class QueueNode:
    def __init__(self, dataval = None):
        self.data = dataval;
        self.next = None;

class Queue:
    def __init__(self):
        self.head = None;
    
    def add(self, value):
        if(self.head == None):
            self.head = QueueNode(value);
        else:
            n = self.head;
            #n = n.next;
            while(n.next != None):
                n = n.next;
            n.next = QueueNode(value);
    
    def remove(self):
        if(self.head == None):
            print("Queue is Empty");
        else:
            top = self.head;
            self.head = self.head.next;
            top.next = None;
            print("Value Removed: ", top.data);
    
    def peek(self):
        if(self.head == None):
            print("Queue is empty.");
        else:
            print("Value at the top of the queue: ", self.head.data);
    
    def isEmpty(self):
        if(self.head == None):
            print("Queue is empty");
        else:
            print("Queue is not empty.");
            
    def printQueue(self):
        if(self.head == None):
            print("Queue is Empty");
        else:
            n = self.head;
            print(n.data);
            n = n.next;
            while(n.next != None):
                print(n.data);
                n = n.next;
            print(n.data);
            
queue1 = Queue();
queue1.peek();
queue1.remove();
queue1.printQueue();
queue1.add(5);
queue1.add(74);
queue1.add(98);
queue1.add(12);
queue1.peek();
queue1.isEmpty();
queue1.printQueue();
queue1.remove();
queue1.printQueue();
queue1.isEmpty();
```

```python
class StackNode:
    def __init__(self, dataval = None, datamin = None):
        self.data = dataval;
        self.minimum = datamin;
        self.next = None;

class StackMod:
    def __init__(self):
        self.head = None;
    
    def push(self, value):
        if(self.head == None):
            newNode = StackNode(value, value);
            self.head = newNode;
        else:
            n = self.head;
            minVal = n.minimum;
            if (value> minVal):
                newNode = StackNode(value, minVal);
                self.head = newNode;
                newNode.next = n;
            else:
                newNode = StackNode(value, value);
                self.head = newNode;
                newNode.next = n;
    
    def pop(self):
        if(self.head == None):
            print("Stack is Empty");
        else:
            n = self.head;
            self.head = self.head.next;
            n.next = None;
            print("Value Removed: ", n.data);
    
    def minStack(self):
        if(self.head ==  None):
            print("Stack is Empty. No minimum value.");
        else:
            print("Minimum value in stack is: ", self.head.minimum);
            
    def printStack(self):
        if(self.head == None):
            print("Stack is empty.");
        else:
            n= self.head;
            print(n.data);
            while(n.next != None):
                n = n.next;
                print(str(n.data) + "    "+  str(n.minimum));


stack2 = StackMod();
stack2.pop();
stack2.minStack();
stack2.push(10);
stack2.push(5);
stack2.push(15);
stack2.push(23);
#stack2.push(3);
stack2.pop();
stack2.pop();
stack2.pop();
stack2.pop();
stack2.pop();
stack2.minStack();
stack2.printStack();
```
